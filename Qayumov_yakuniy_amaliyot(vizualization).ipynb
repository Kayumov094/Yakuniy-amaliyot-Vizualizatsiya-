Hukumatlar, tashkilotlar va fuqarolik jamiyati oʻz siyosati, qarorlarini qabul qilishda baxtiyorlik koʻrsatkichlaridan tobora koʻproq foydalanmoqda. Shu sabab hisobot jahon miqyosida eʼtirof etilishda davom etmoqda. Iqtisodiyot, psixologiya, milliy statistika, sog‘liqni saqlash, davlat siyosati va boshqa sohalardagi yetakchi ekspertlar xalqlar taraqqiyotini baholashda farovonlik o‘lchovlaridan qanday samarali foydalanish mumkinligini tasvirlab beradi. Hisobotlar bugungi kunda dunyodagi baxtiyorlik darajasining holatini koʻrsatadi.

---

### **Ustunlar tavsifi**  

1. **Country** – mamlakat nomi  
   - Bu satrda mamlakatlarning nomlari joylashgan.  

2. **Happiness.Rank** – baxtlilik darajasi bo‘yicha reyting  
   - Har bir mamlakatning baxt darajasi bo‘yicha egallagan o‘rni.  
   - **1** – eng baxtli mamlakat, **so‘nggi o‘rin** – eng baxtsiz mamlakat.  

3. **Happiness.Score** – baxtlilik darajasi  
   - Mamlakatning umumiy baxtlilik darajasi **(0–10 oralig‘ida)**.  
   - Bu baho turli omillarga asoslangan holda berilgan.  

4. **Whisker.high** – yuqori ishonch oralig‘i (Upper Confidence Interval)  
   - **Happiness Score** qiymatining yuqori chegarasi (ishonchlilik darajasi bilan).  

5. **Whisker.low** – pastki ishonch oralig‘i (Lower Confidence Interval)  
   - **Happiness Score** qiymatining pastki chegarasi.  

6. **Economy..GDP.per.Capita.** – iqtisodiy farovonlik (YAIM)  
   - Aholi jon boshiga to‘g‘ri keladigan yalpi ichki mahsulot (YAIM).  
   - Mamlakat iqtisodining baxtlilik darajasiga ta’sirini ko‘rsatadi.  

7. **Family** – jamiyatning qo‘llab-quvvatlashi  
   - Ijtimoiy aloqalar va oilaviy qo‘llab-quvvatlash darajasi.  
   - Ushbu qiymat qanchalik yuqori bo‘lsa, odamlar o‘zini shunchalik baxtli his qiladi.  

8. **Health..Life.Expectancy.** – sog‘lom umr davomiyligi  
   - Odamlarning o‘rtacha qancha umr ko‘rishi va sog‘lom hayot kechirishi darajasi.  
   - Tibbiy xizmat sifati va hayot sharoitlari bilan bog‘liq.  

9. **Freedom** – hayot tanlovlarida erkinlik  
   - Odamlarning o‘z hayotini erkin boshqarish darajasi.  
   - Mamlakatdagi demokratik va fuqarolik erkinliklarini aks ettiradi.  

10. **Generosity** – saxiylik darajasi  
   - Jamiyatdagi saxiylik va xayriya qilishga moyillik darajasi.  
   - Ushbu qiymat qanchalik yuqori bo‘lsa, odamlar ko‘proq yordam berishga tayyor bo‘ladi.  

11. **Trust..Government.Corruption.** – korrupsiya haqidagi tasavvurlar  
   - Mamlakatdagi korrupsiya darajasi bo‘yicha odamlarning fikri.  
   - **0 ga yaqin bo‘lsa**, odamlar davlatga ishonmaydi.  
   - **Yuqori qiymat bo‘lsa**, davlatga ishonch baland bo‘ladi.  

12. **Dystopia.Residual** – baxtiyorlik darajasiga ta’sir etmaydigan omillar  
   - Ideal bo‘lmagan (Dystopia) sharoitlarda baxtlilik darajasini oshirishga yordam beruvchi faktorlar.  
   - Bu ko‘rsatkich mamlakat baxtlilik darajasi **modelining qoldiq qiymati** sifatida tushuniladi.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/happines_rank_2017.csv"
df = pd.read_csv(url)
df.head(3)
```

#1) Baxtlilik darajasi va aholi jon boshiga to‘g‘ri keladigan YAIM o‘rtasidagi bog‘liqlikni scatterplot bilan vizualizatsiya qiling.

```python
#javobni shu yerda yozing.
df['Economy..GDP.per.Capita.'] = pd.to_numeric(df['Economy..GDP.per.Capita.'], errors='coerce')
df['Happiness.Score'] = pd.to_numeric(df['Happiness.Score'], errors='coerce')
```

```python
plt.figure(figsize =(12,6))
sns.scatterplot(data = df, x = "Economy..GDP.per.Capita.", y = "Happiness.Score")
plt.title("Baxtlilik darajasi va YAIM o'rtasidagi bog'liklik", color = 'b', fontsize = 16)
plt.xlabel("Aholi jon boshiga YAIM", fontsize = 12)
plt.ylabel("baxtlilik darajasi", fontsize = 12)
plt.grid(True)
plt.show()
```

#2) Baxtlilik darajasi (`Happiness.Score`) taqsimotini histogram orqali chizing.

```python
#javobni shu yerda yozing.
plt.figure(figsize = (10,6))
plt.title("Baxtlilik darajasining taqsimoti")
sns.histplot(data = df, x = "Happiness.Score", kde = True,  color = 'lightblue')
plt.xlabel("Baxtlilik darajasining taqsimoti")
plt.ylabel("Davlatlar soni")
plt.grid(True)
plt.show()
```

#3) Eng baxtli va eng baxtsiz 10 ta mamlakatni barplot orqali ko‘rsating.

```python
#javobni shu yerda yozing.
df_baxtli = df.head(10)
df_baxtsiz = df.tail(10)
df_baxtli
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,10)) # subplotlar yaratib oldim
plt.suptitle("Eng baxtli va eng baxtsiz davlatlar ko'rsatlichlari", fontsize=26)
# Top 10 davlat uchun grafik
sns.barplot(ax = axes[0],
            data = df_baxtli,
            x = 'Happiness.Score',
            y = 'Country',
            hue ='Whisker.high',
            palette='Greens')
axes[0].set_title("Eng baxtli davlatlat o'ntaligi", fontsize=22)
axes[0].set_xlabel("Baxtlilik darajasi", fontsize=18)
axes[0].set_ylabel("Top o'nta davlat", fontsize=18)

# Bottom 10 davlat uchun grafik
df_baxtsiz_sorted = df_baxtsiz.sort_values(by='Happiness.Score', ascending=True)
sns.barplot(ax = axes[1],
            data = df_baxtsiz_sorted,
            x ='Happiness.Score',
            y ='Country',
            hue ='Whisker.high',
            palette='Blues')
axes[1].set_title("Eng baxtsiz davlatlat o'ntaligi", fontsize=22)
axes[1].set_xlabel("Baxtlilik darajasi", fontsize=18)
axes[1].set_ylabel("Quyi o'nta davlat", fontsize=18)

plt.tight_layout()
plt.show()
```

#4)  Baxtlilik darajasi (`Happiness.Score`) va ijtimoiy omillar (Freedom, Family Support, Generosity) bilan scatterplot (`subplot` orqali ketma ket) chizing!

```python
fig, axes = plt.subplots(1,3, figsize=(18,6),sharey = True)
plt.suptitle("Baxlilik darajasining ijtimoiy omillar bilan bog'lik darajasi", fontsize = 24)

# Uchta subplotga plot chizamiz
sns.scatterplot(ax = axes[0],data = df, y = 'Happiness.Score', x = 'Freedom', color = 'blue')
sns.scatterplot(ax = axes[1], data = df, y = 'Happiness.Score',x = 'Family', color = 'green')
sns.scatterplot(ax = axes[2], data = df, y = 'Happiness.Score',x = 'Generosity', color = 'red')

# Uchta subplotga nom beramiz
axes[0].set_title("Baxtlilik darajasini Freedomga bog'liqligi", fontsize = 16, color = 'r')
axes[1].set_title("Baxtlilik darajasini Familyga bog'liqligi", fontsize = 16, color = 'r')
axes[2].set_title("Baxtlilik darajasini Generosityga bog'liqligi", fontsize = 16, color = 'r')

plt.tight_layout()
plt.show()
```

#5) Korrupsiya va baxtlilik darajasi bog‘liqligini koʻrsating!

```python
#javobni shu yerda yozing.
plt.figure(figsize = (12,6))
sns.scatterplot(data = df, x = df['Happiness.Score'], y=df['Trust..Government.Corruption.'])
plt.title("Korrupsiyaning baxtlilik darajasi bilan bog'liqlik holati", color = 'r')
plt.xlabel("Baxtlilik ko'rsatkichi")
plt.ylabel("Korrupsiya haqida tasavvurlar")
plt.grid(True)
plt.show()
```

#6) Baxtlilik darajasi va YAIM o‘rtasidagi bog‘liqlikni chiziqli regressiya orqali tekshiring!

```python
plt.figure(figsize= (12,6))
sns.regplot(data = df, x = df['Happiness.Score'], y = df['Trust..Government.Corruption.'])
plt.title("Baxtlilik darajasi va YAIM o'rtasidagi chiziqli regressiya", color= 'b', fontsize = 20)
plt.xlabel("Korrupsiya bo'yicha xavotirlar", color = 'b', fontsize = 16)
plt.ylabel("Baxtlilik darajasi", color='b', fontsize='16')

plt.show()
```

#8) 10 ta eng yuqori va eng past YAIMga ega mamlakatlarni barplot orqali aks ettiring!

```python
df_top = df.head(10)
df_bot = df.tail(10)
df_bot = df_bot.sort_values(by = 'Economy..GDP.per.Capita.', ascending = False)
df_bot
```

```python
df_top = df.head(10)
df_bot = df.tail(10)
fig, axes = plt.subplots(1,2, figsize = (14,8))
fig.suptitle("Eng yuqori va eng quyi o'nta davlatning YAIM ko'rsatkichlari",color = 'r', fontsize = 24)
# birinchi barplotga plot chizamiz
sns.barplot(ax = axes[0], data = df_top, color = 'green', x = 'Economy..GDP.per.Capita.', y = 'Country')
axes[0].set_title("Eng yuqori o'nta davlatning YAIM dagi ulushi", fontsize =18)
axes[0].set_ylabel("Davlatlar nomi")
axes[0].set_xlabel("YAIM")

# ikkinchi barplotga plot chizamiz
sns.barplot(ax = axes[1], data = df_bot, color = 'b', x = 'Economy..GDP.per.Capita.', y = 'Country')
axes[1].set_title("Eng quyi o'nta davlatning YAIM dagi ulushi", fontsize =18)
axes[1].set_xlabel("YAIM")

plt.tight_layout()
plt.show()
```

#9) Oilaviy qo‘llab-quvvatlash va YAIM o‘rtasidagi bog‘liqlikni aks ettiring!

```python
fig, ax = plt.subplots(1,1, figsize = (12,6), facecolor = '#7BC8F6' )
sns.regplot(ax = ax, data = df, x = 'Family', y = 'Economy..GDP.per.Capita.', color = "b")
ax.set_title("Mamlakatlarning oilaviy qo'llab-quvvatlash bilan\n YAIM o'rtasidagi bog'liqlik",color='b', fontsize = 24)
ax.set_xlabel("Oilaviy qo'llab-quvvatlash",color = 'b', fontsize = '14')
ax.set_ylabel("YAIM", color = 'b', fontsize = '14')
ax.set_facecolor("#00FFFF")
sns.despine(right = True, top = True)
plt.tight_layout()
plt.show()
```


# 10)  Saxiylik (Generosity) va hukumat korrupsiyasi (Trust..Government.Corruption.) o‘rtasidagi bog‘liqlikni scatterplot yordamida vizualizatsiya qiling!

```python
fig, ax = plt.subplots(1,1, figsize = (10,6))
sns.scatterplot(ax = ax, data = df, x = 'Generosity', y = 'Trust..Government.Corruption.', color = 'b')
ax.set_title("Saxiylik va korrupsiya o'rtasidagi bog'liqlik", color = 'b', fontsize = 24)
ax.set_xlabel("Saxiylik darajasi", color = 'b', fontsize = 14)
ax.set_ylabel("Korrupsiya haqida tasavvurlar", color = 'b', fontsize =14)
sns.despine(right = True, top = True)
plt.tight_layout()
plt.show()
```

#11) Baxtlilik darajasi bo‘yicha mamlakatlarni “Juda baxtli”, “Baxtli”, “O‘rtacha”, “Baxtsiz” toifalariga ajrating va har bir guruh uchun o‘rtacha iqtisodiy ko‘rsatkichlarni (GDP per Capita) ko‘rsating.

```python
df["Happiness_guruhlari"]  = pd.qcut(df["Happiness.Rank"], q=4, labels = ["Juda baxtli", "Baxtli", "O'rtacha", 'Baxtsiz'])
df.groupby("Happiness_guruhlari")["Economy..GDP.per.Capita."].mean()
sns.barplot(data = df, x = 'Happiness_guruhlari', y = 'Economy..GDP.per.Capita.', color = '#ADD8E6',hue = 'Whisker.high')
plt.title("Guruhlar bo'yicha O'rtacha YAIM ko'rsatkichlari", color = 'b')
plt.xlabel("Guruhlar")
plt.ylabel("O'rtacha YAIM")
sns.despine(right = True, top = True)
plt.tight_layout()
plt.show()
```

#12) Muhim o‘zgaruvchilar (GDP, Life Expectancy, Freedom, Generosity, Happiness Score) bo‘yicha pairplot orqali umumiy taqsimotni vizualizatsiya qiling.

```python
muhimlar = ['Economy..GDP.per.Capita.', 'Health..Life.Expectancy.', 'Freedom', 'Generosity', 'Happiness.Score']
sns.pairplot(data = df[muhimlar], diag_kind = 'kde')
plt.show()
```
